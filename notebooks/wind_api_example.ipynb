```python
# Apply date rules example
# Rule format: +/-<number><symbol>
# Symbols: B=business day, M=month, W=week, d=day, Q=quarter, Y=year

base_date = '2024-01-15'

# Get 5 business days before
result = apply_date_rule(base_date, '-5B', 'CBT')
print(f"5 business days before {base_date}: {result}")

# Get 1 month ahead
result = apply_date_rule(base_date, '+1M', 'CBT')
print(f"1 month after {base_date}: {result}")

# Apply multiple rules
result = apply_date_rules(base_date, '+1M-5B', 'CBT')
print(f"1 month minus 5 business days from {base_date}: {result}")

result = apply_date_rules(base_date, '-2Y', 'CBT')
print(f"2 years before {base_date}: {result}")
```

## 5. Query Data Using Wind API

```python
# Import Wind API components
from data_api.wind import WindDownloader, WindPipeline, WindConfig

# Create configuration (uses default fields from WindConfig)
config = WindConfig(data_path='data/wind')

# Display default fields for futures
print("Default fields for futures (from WindConfig):")
print(f"  {config.future_fields}")
print()

# Create downloader and pipeline
downloader = WindDownloader(config.to_dict())
pipeline = WindPipeline(downloader, config.to_dict())
```

```python
# Connect to Wind Terminal
# NOTE: Wind Terminal must be running for this to work

if downloader.connect():
    print("Successfully connected to Wind Terminal")
else:
    print("Failed to connect to Wind Terminal. Make sure Wind Terminal is running.")
```

```python
# Query data for Pre-2020 contracts
# Using fields=None will use the default future_fields from WindConfig

print("=" * 60)
print("Querying Pre-2020 Contract Data (2 months before expiration)")
print("=" * 60)

# DCE No.1 Soybean - Pre-2020
print(f"\nQuerying {dce_pre2020_ticker} from {dce_pre2020_start} to {dce_pre2020_ltd}")
dce_pre2020_result = pipeline.run_single(
    symbol=dce_pre2020_ticker,
    start_date=dce_pre2020_start,
    end_date=dce_pre2020_ltd,
    fields=None,  # Uses default future_fields
    save_to_disk=False
)

if dce_pre2020_result['success']:
    print(f"  Success! Data shape: {dce_pre2020_result['data'].shape}")
    print(f"  Columns: {list(dce_pre2020_result['data'].columns)}")
    display(dce_pre2020_result['data'].head())
else:
    print(f"  Failed: {dce_pre2020_result['error']}")
```

```python
# CZCE White Sugar - Pre-2020
print(f"\nQuerying {czce_pre2020_ticker} from {czce_pre2020_start} to {czce_pre2020_ltd}")
czce_pre2020_result = pipeline.run_single(
    symbol=czce_pre2020_ticker,
    start_date=czce_pre2020_start,
    end_date=czce_pre2020_ltd,
    fields=None,  # Uses default future_fields
    save_to_disk=False
)

if czce_pre2020_result['success']:
    print(f"  Success! Data shape: {czce_pre2020_result['data'].shape}")
    print(f"  Columns: {list(czce_pre2020_result['data'].columns)}")
    display(czce_pre2020_result['data'].head())
else:
    print(f"  Failed: {czce_pre2020_result['error']}")
```
