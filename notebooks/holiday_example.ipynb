# Holiday Calendar Example

This notebook demonstrates how to use the `my_holiday` package to work with holiday calendars.

```python
# Import CBT (Chicago Board of Trade) holiday calendar
from my_holiday import CBT

# Create a CBT holiday calendar instance
cbt_holidays = CBT()

# Display the holiday calendar
print("CBT Holiday Calendar:")
print(cbt_holidays)
```

```python
# Import utility functions for date calculations
from my_holiday import (
    is_holiday,
    next_business_day,
    previous_business_day,
    business_days_between,
    apply_date_rule,
    apply_date_rules
)

# Check if a date is a holiday
date_to_check = '2024-12-25'
print(f"Is {date_to_check} a holiday? {is_holiday(date_to_check, 'CBT')}")

# Get the next business day
next_biz = next_business_day('2024-12-25', 'CBT')
print(f"Next business day after 2024-12-25: {next_biz}")
```

```python
# Apply date rules example
# Rule format: +/-<number><symbol>
# Symbols: B=business day, M=month, W=week, d=day, Q=quarter, Y=year

base_date = '2024-01-15'

# Get 5 business days before
result = apply_date_rule(base_date, '-5B', 'CBT')
print(f"5 business days before {base_date}: {result}")

# Get 1 month ahead
result = apply_date_rule(base_date, '+1M', 'CBT')
print(f"1 month after {base_date}: {result}")

# Apply multiple rules
result = apply_date_rules(base_date, '+1M-5B', 'CBT')
print(f"1 month minus 5 business days from {base_date}: {result}")
```
